```python
import numpy as np  
from scipy.stats import norm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

sample_count = 30

experiment = np.random.randint(-11, 12, sample_count)
type(experiment)
experiment


X = np.arange(-11, 12, 1)
Y = np.arange(1, 12, 1)
X1, Y1 = np.meshgrid(X, Y)
m = norm (loc = 0, scale = 9)
m.pdf(experiment)
np.prod(m.pdf(experiment))
d = m.pdf(X1)*m.pdf(Y1)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X1, Y1, d, rstride=1, cstride=1, cmap='coolwarm')
plt.show()
np.sum(d)
for i in range(5):
    for ii, x in enumerate(X):
        for jj, y in enumerate(Y):
            m_ = norm (loc = x, scale = y)
            l = np.prod(m_.pdf(experiment))
            d[jj][ii] *= l
    d /= np.sum(d)
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_surface(X1, Y1, d, rstride=1, cstride=1, cmap='coolwarm')
    plt.show()
```
